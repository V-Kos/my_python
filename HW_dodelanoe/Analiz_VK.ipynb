```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dm-fedorov/python_basic/master/data/vkontakte_group_01_2016-08-01_2020-03-15.csv")
df.head(50)
df['Дата'] =pd.to_datetime(df.Дата)
df=df.sort_values(by=['Дата'])
df.head(50)
```
